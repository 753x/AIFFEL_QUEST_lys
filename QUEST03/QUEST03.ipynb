```python
import os
import numpy as np


file_path = '/content/drive/MyDrive/Avengers/Avengers.txt'                      # 파일을 드라이브에서 불러옴.

def read_file(file_path):                                                       # 파일을 읽기 파일로 열고 content에 저장
    with open(file_path, 'r') as file:
        content = file.readlines()
    return content                                                              # content로 반환

def split_string(content):                                                      # 줄바꿈
    lines = [line.strip() for line in  content]
    return lines

def split_space(content) :                                                      # 공백 제거
    space_lines = []
    for line in content:
        words = line.split(' ')
        space_lines.extend(words)
    space_lines = [item for item in space_lines if item]  # 빈 문자열 제거
    return space_lines

def lowercase_strings(lines):                                                   # 소문자로 변환
    lowercase_lines = [line.lower() for line in lines]
    return lowercase_lines

def remove_symbols(text):                                                       # 기호 제거
    symbols = "~`!@#$%^&*()-_+={}[]|\\:;'<>?,./\""
    for symbol in symbols:
        text = text.replace(symbol, "")
    return text

def get_2gram(text):                                                            #2-gram 활용
    ngrams = []
    for i in range(len(text) - 1):
        ngrams.append(text[i:i+2])
    return ngrams



content = read_file(file_path)
lines = split_string(content)
lowercase_lines = lowercase_strings(lines)
space_lines = split_space(lowercase_lines)

# 기호 제외하고 2-gram으로 나누어 저장
processed_lines = []
for line in space_lines:
    line = remove_symbols(line)
    ngrams = get_2gram(line)
    processed_lines.extend(ngrams)

print(processed_lines)

from collections import Counter

# processed_lines에서 2-gram 페어의 빈도를 계산
ngram_counter = Counter(processed_lines)

# 가장 빈도가 높은 2-gram 페어를 찾음
most_common_pair = ngram_counter.most_common(1)

print(most_common_pair)
```
